```python
import pandas as pd

file_path = './data/task.csv'

df = pd.read_csv(file_path)
print(df)
df.head(10)
```

```python
df.groupby(by=['Компонент'])['Затрачено в часах'].mean().to_frame().head()
```

```python
import numpy as np

agg_config = {'Затрачено в часах':{
    'mean_val': np.mean, 'max_val':np.max}
}

df.groupby(by=['Компонент']
)[
    ['Затрачено в часах']
].aggregate(
    agg_config
)
```
