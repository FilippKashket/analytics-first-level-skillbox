```python
import pandas as pd

text_df = pd.read_csv("./data-2/content_description.csv", sep='\t')
text_df.head()
```

```python
import string

sample_text = text_df.description.values[1]

print(sample_text + '\n\n')

print("Нижний регистр")
print(sample_text.lower()+'\n\n')

print("Без знаков препинания")
print("".join([i for i in sample_text.lower() if i not in string.punctuation])+'\n\n')
```
