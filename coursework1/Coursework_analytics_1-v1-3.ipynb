```python
import datetime


df_increase=df_main[['course_id','course_title', 'student_id', 'start_date', 'is_homework']]
df_increase = df_increase[(df_increase['is_homework']==True)]
df_increase['start_month'] =df_increase['start_date'].apply(lambda date: datetime.datetime.strptime(date[0:7], '%Y-%m'))

df_increase=df_increase.groupby(['course_id','course_title', 'student_id'])['start_month'].min().reset_index()

df_increase['dummy'] =1

df_increase = df_increase.groupby(['course_id','course_title','start_month'])['dummy'].sum().reset_index()
 
df_increase = df_increase.sort_values(by='start_month')

df_increase
```

```python
import matplotlib.pyplot as plt

courses_list = df_increase.groupby(['course_title', 'course_id'])['dummy'].sum().reset_index()

for index, row in courses_list.iterrows():
    fig, subplot = plt.subplots()
    fig.set_size_inches(9, 5, forward=True)
    course_data = df_increase[(df_increase['course_id']==row.course_id)]
    
    start_date = course_data['start_month'].min()
    end_date = course_data['start_month'].max()
    
    m = pd.date_range(start_date, end_date, freq='MS')
    
    for month in m:
        if(course_data[(course_data['start_month']==month)]['start_month'].size==0):
            course_data.loc[month]= [row.course_id,row.course_title,month, 0]
    
    course_data = course_data.sort_values(by='start_month')
    subplot.set_title(row.course_title)
    subplot.plot(course_data['start_month'], course_data['dummy'])
    
    subplot.xaxis.set_ticks(m.values)
    
    subplot.set_ylabel('Число новых струдентов')
    subplot.set_xlabel('месяц')
    plt.xticks(rotation='vertical')

plt.show()
```
