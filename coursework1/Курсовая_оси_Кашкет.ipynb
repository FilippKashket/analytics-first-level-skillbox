```python
import pandas as pd

df_main_1 = df_main.copy()

df_main_1 = df_main_1.dropna()
df_main_1['start_date'] = pd.to_datetime(df_main_1['start_date'])
df_main_1['finish_date'] = pd.to_datetime(df_main_1['finish_date'])

df_main_1 = df_main_1[(df_main_1['finish_date'] > df_main_1['start_date'])]

df_result = df_main_1.groupby(['course_id', 'course_title', 'module_number','progress_id']).agg({'start_date': 'min', 'finish_date':'max'}).reset_index()
df_result['difference'] = (df_result['finish_date'] - df_result['start_date']).dt.total_seconds()
df_result = df_result.groupby(['course_id', 'course_title', 'module_number'])['difference'].agg({'min','max','median','mean'}).reset_index()

df_result
```

```python
import matplotlib.pyplot as plt

courses_list = df_main.groupby(["course_id", "course_title"])['module_number'].nunique().reset_index()

for index, row in courses_list.iterrows():
    fig, subplot = plt.subplots()
    fig.set_size_inches(10.5,6.5 , forward=True)
    subplot.grid()
    course_data = df_result[(df_result['course_id'] == row.course_id)]
    subplot.set_title(row.course_title)
    subplot.plot(course_data['module_number'], course_data['mean']) 
    subplot.xaxis.set_ticks(range(0, 10, 1))
    
    subplot.set_ylabel('Медианное время завершения домашней работы')
    subplot.set_xlabel('Номер модуля')
    plt.xticks(rotation='vertical')

plt.show()
```

```python
import pandas as pd

df_main_2 = df_main.copy()

df_main_2 = df_main_2.dropna()
df_main_2['finish_month'] = df_main_2['finish_date'].apply(lambda x: x[0:7]) 

df_main_2['start_date'] = pd.to_datetime(df_main_2['start_date'])
df_main_2['finish_date'] = pd.to_datetime(df_main_2['finish_date'])

df_main_2 = df_main_2[(df_main_2['finish_date'] > df_main_2['start_date'])]

df_result = df_main_2.groupby(['course_id', 'course_title', 'finish_month','progress_id']).agg({'start_date': 'min', 'finish_date':'max'}).reset_index()
df_result['difference'] = (df_result['finish_date'] - df_result['start_date']).dt.total_seconds()
df_result = df_result.groupby(['course_id', 'course_title', 'finish_month'])['difference'].agg({'median'}).reset_index()

df_result['finish_month'] = pd.to_datetime(df_result['finish_month'])

df_result
```

```python
import matplotlib.pyplot as plt

courses_list = df_main.groupby(["course_id", "course_title"])['module_number'].nunique().reset_index()

fig, subplot = plt.subplots()
fig.set_size_inches(18.5, 10.5, forward=True)
subplot.grid()
for index, row in courses_list.iterrows():
    course_data = df_result[(df_result['course_id'] == row.course_id)] 
    
    subplot.plot(course_data['finish_month'], course_data['median'])
plt.legend(courses_list['course_title'])
plt.xticks(rotation=45)
plt.show()
```
